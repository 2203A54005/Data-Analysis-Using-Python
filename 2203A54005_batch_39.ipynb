```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load datset
data = pd.read_csv('/content/insurance.csv')
print(data)
print(data.head())
print(data.columns)
```

```python
#Create vectors, factors, lists, matrices, and data frames.
# Vector
age_vector = data['age'].values

# Factor (Categorical in Python)
smoker_factor = pd.Categorical(data['smoker'])

# List
data_list = [data['age'], data['smoker'], data['expenses']]

# Matrix (2D Array in Python)
matrix = np.array([data['age'], data['expenses']])
#data frame
new_data_frame = pd.DataFrame({
    'Age': data['age'],
    'Smoking_Status': data['smoker'],
    'Expenses': data['expenses']
})
print(new_data_frame.head())

```

```python
#b) Generate a scatter plot. Customize it by adding a title, axis labels, text annotations, and changing the plot characters (e.g., circles to plus signs).
#c) Filter the data based on a condition and updates the plot to reflect the filtered data.

#Plot a scatter plot between age and smoking,
plt.figure(figsize=(6, 6))
plt.scatter(data['age'], data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='blue', alpha=0.6, label='Smoking Status')
plt.title('Scatter Plot: Age vs Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
# Plot a scatter plot between age(20 to 40) and smoking,
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(6, 6))
plt.scatter(filtered_data['age'], filtered_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='green', alpha=0.6, label='Smoking Status')
plt.title('Scatter Plot: Age (20-40) vs Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
#Find expenses of smokers and non smokers,(plot)
smoker_expenses = data[data['smoker'] == 'yes']['expenses']
non_smoker_expenses = data[data['smoker'] == 'no']['expenses']
plt.figure(figsize=(6, 6))
plt.boxplot([smoker_expenses, non_smoker_expenses], labels=['Smokers', 'Non-Smokers'], patch_artist=True)
plt.title('Expenses: Smokers vs Non-Smokers')
plt.ylabel('Expenses')
plt.grid(alpha=0.3)
plt.show()
```

```python
#Find the number of smokers gender wise (plot)
smoker_gender_counts = data[data['smoker'] == 'yes']['sex'].value_counts()
plt.figure(figsize=(6, 6))
smoker_gender_counts.plot(kind='bar', color=['grey', 'pink'])
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Smokers')
plt.xticks(rotation=0)
plt.grid(alpha=0.3)
plt.show()
```
