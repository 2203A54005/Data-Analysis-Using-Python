```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Question 1(a): Binomial Probability for Smokers
n = 10  # sample size
p = 0.3  # probability of a smoker
x = np.arange(4, 11)  # At least 4 smokers
prob_at_least_4 = 1 - binom.cdf(3, n, p)
print(f"Probability of at least 4 smokers: {prob_at_least_4:.4f}")
```

```python
# Question 1(b): Plot Binomial Distribution
x_values = np.arange(0, n+1)
pdf_values = binom.pmf(x_values, n, p)
plt.bar(x_values, pdf_values, color='blue', alpha=0.6)
plt.xlabel('Number of smokers')
plt.ylabel('Probability')
plt.title('Binomial Distribution of Smokers')
plt.show()
```

```python
# Question 2(a): Risk Estimation of High Medical Expenses
n2 = 15  # sample size
p2 = 0.2  # probability of high expenses
prob_exactly_5 = binom.pmf(5, n2, p2)
print(f"Probability of exactly 5 high-expense individuals: {prob_exactly_5:.4f}")
```

```python
# Question 2(b): Cumulative Probability at most 5
prob_at_most_5 = binom.cdf(5, n2, p2)
print(f"Probability of at most 5 high-expense individuals: {prob_at_most_5:.4f}")
```

```python
# Question 2(c): Plot CDF
cdf_values = binom.cdf(x_values, n2, p2)
plt.plot(x_values, cdf_values, marker='o', linestyle='--', color='red')
plt.xlabel('Number of individuals with high expenses')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Probability Distribution')
plt.show()
```

```python
# Question 3(a): Simulating Smoking Rates in Regions
regions = {'Southwest': 0.25, 'Southeast': 0.30, 'Northwest': 0.20, 'Northeast': 0.35}
n3 = 50  # sample size
simulated_data = {region: np.random.binomial(n3, p, 1000) for region, p in regions.items()}
```

```python
# Question 3(b): Visualize distribution per region
plt.figure(figsize=(8, 6))
for region, data in simulated_data.items():
    plt.hist(data, alpha=0.5, label=region, bins=15, density=True)
plt.xlabel('Number of smokers')
plt.ylabel('Frequency')
plt.legend()
plt.title('Simulated Smokers per Region')
plt.show()
```

```python
# Question 3(c): Compute expected smokers
expected_smokers = {region: n3 * p for region, p in regions.items()}
print("Expected number of smokers per region:", expected_smokers)
```

```python
# Question 4
# (a) Find threshold for top 10%
from scipy import stats
n4 = 30
p4 = 0.1  # Top 10%
thresh = stats.binom.ppf(0.9, n4, p4)
print(f"Threshold charge for high-risk individuals: {thresh:.0f}")
```

```python
# Question 5(a): Probability of at least 5 individuals with >3 children
n5 = 12
p5 = 0.4
prob_at_least_5_children = 1 - binom.cdf(4, n5, p5)
print(f"Probability of at least 5 individuals with >3 children: {prob_at_least_5_children:.4f}")
```

```python
# Question 5(b): Simulate 1000 samples and plot histogram
simulated_children = np.random.binomial(n5, p5, 1000)
plt.hist(simulated_children, bins=15, color='purple', alpha=0.6, density=True)
plt.xlabel('Individuals with >3 children')
plt.ylabel('Frequency')
plt.title('Simulated Distribution for Number of Children')
plt.show()
```

```python
# Question 5(c): Compare mean and variance
mean_simulated = np.mean(simulated_children)
variance_simulated = np.var(simulated_children)
mean_theoretical = n5 * p5
variance_theoretical = n5 * p5 * (1 - p5)
print(f"Simulated Mean: {mean_simulated:.2f}, Theoretical Mean: {mean_theoretical:.2f}")
print(f"Simulated Variance: {variance_simulated:.2f}, Theoretical Variance: {variance_theoretical:.2f}")
```
